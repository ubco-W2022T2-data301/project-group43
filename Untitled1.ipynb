```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df=df.drop(columns=['Ranked2021','Ranked2020','Country','Sector'])
df
```

```python
sns.scatterplot(x='Employees2020',y='CAGR',data=df)
plt.xlabel('Number of Employees 2020')
plt.ylabel('CAGR')
```
